Question: What are the top games currently and who are the top streamers of those top games?
Business Justification: It is valuable to know the top 5 games that have the most viewer traffic. In addition, top streamers are valuable to know because of their influencer value. Top twitch streamers can heavily influence the viewer traffic of Twitch
SQL features used: Get, JOIN, VIEWS, GROUP BY

```python
%%sql
SELECT 
    sv.stream_rank_daily,
    sv.user_name,
    sv.game_name,
    tg.pop_game,
    sv.game_id,
    sv.viewer_count
FROM Stream_viewer sv
JOIN TopGames tg
    ON sv.game_id = tg.id
WHERE pop_game < 6
GROUP BY game_name 
ORDER BY pop_game, viewer_count DESC
```

Insights: Just Chatting, GTA5, League of Legends, Fortnite, and FIFA 22 are the top 5 games currently. Things to note is that even though GTAV is number 2 in pop_game, their viewer_count is very low due to the fact that viewer_count is calculated by only the top 100

Question: Who are the streamers streaming the top three games, what is their viewer count, and what is the total viewers for each game
Business Justification: The twitch api doesn't have a total viewer count call. Having a total viewer count on a particular game allows for futher analysis.
SQL features used: Common Table Expression (CTE), JOINS, Window Function

```python
%%sql
SELECT 
    sv.stream_rank_daily,
    sv.user_name,
    sv.game_name,
    tg.pop_game,
    sv.game_id,
    sv.viewer_count,
    SUM(viewer_count) OVER (PARTITION BY game_name ORDER BY tg.pop_game) AS game_viewer_total
FROM Stream_viewer sv
JOIN TopGames tg
    ON sv.game_id = tg.id
WHERE pop_game < 4
ORDER BY tg.pop_game
```

Insights: While GTA is the 2nd most popular game by viewership, it loses out to League of Legends in viewer_total of the top 100 streamers at the time. This is interesting because it means GTA viewership is carried by many smaller streamers rather than the top 100 streamers by viewer.

Question: What percentage of viewership do each streamer take in regards to the game they are playing? 
Business Justification: Based on the past quary getting the percentage of viewership would allow to visualize how spread out viewship is across multiple streamers. Streamers with too high of a viewship percentage of a game can be problematic for smaller streamers who are trying to gain footholds in viewship.
SQL functions used: CTE, JOINS, Window Functions, CASE

The first quary initally found the percentage of viewership which lead to the idea of pointing out/determing which streamers oversaturate a viewership of a particular game. Any streamer that has more than 20% of the total viewership is flagged with oversaturation of a game. 

```python
%%sql
WITH game_total_views AS(
    SELECT 
        sv.stream_rank_daily,
        sv.user_name,
        sv.game_name,
        tg.pop_game,
        sv.game_id,
        sv.viewer_count,
        SUM(viewer_count) OVER (PARTITION BY game_name ORDER BY tg.pop_game) AS game_viewer_total
    FROM Stream_viewer sv
    JOIN TopGames tg
        ON sv.game_id = tg.id
    WHERE pop_game < 5
    ORDER BY tg.pop_game, viewer_count DESC
)
    SELECT
        user_name,
        game_name,
        viewer_count,
        game_viewer_total,
        viewer_count / game_viewer_total * 100 AS percentage_of_viewership
    FROM game_total_views
```

```python
%%sql
WITH game_total_views AS(
    SELECT 
        sv.stream_rank_daily,
        sv.user_name,
        sv.game_name,
        tg.pop_game,
        sv.game_id,
        sv.viewer_count,
        SUM(viewer_count) OVER (PARTITION BY game_name ORDER BY tg.pop_game) AS game_viewer_total
    FROM Stream_viewer sv
    JOIN TopGames tg
        ON sv.game_id = tg.id
    WHERE pop_game < 5
    ORDER BY tg.pop_game, viewer_count DESC
),
percentage_viewership AS(
    SELECT
        user_name,
        game_name,
        viewer_count,
        game_viewer_total,
        viewer_count / game_viewer_total * 100 AS percentage_of_viewership
    FROM game_total_views
)
SELECT
    user_name,
    game_name,
    viewer_count,
    game_viewer_total,
    percentage_of_viewership,
    CASE 
        WHEN percentage_of_viewership > 20 THEN 'Oversaturated'
        ELSE 'Undersaturated'
    END AS game_saturation_check
FROM percentage_viewership
```

Insight: Games are fairly evenly spreadout across all streamers within their games. There is a main expection whcih is loltyler1 who takes nearly 70% of the league of legends viewship percentage. This domination in viewership causes more and more viewers to join when browsing through League of Legends which can be valuable in terms of advertisement / influences likes Twitch drops. It is highly likely that loltyler1 regularlly attracts his loyal followers.
